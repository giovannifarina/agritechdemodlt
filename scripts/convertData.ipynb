```python
import datetime
import time

with open('../tracciaConverted.csv','w') as ofd:
    with open('../traccia.csv','r') as fd:
        fd.readline()
        for line in fd:
            tokens = line.split(",")
            timestamp, lat, long = tokens[0], tokens[2],tokens[1]
            date_time = datetime.datetime.strptime(timestamp, "%d-%b-%Y %H:%M:%S")
            unix_timestamp = int(time.mktime(date_time.timetuple()))
            print(unix_timestamp,lat,long)
            ofd.write(f"{unix_timestamp},{lat},{long}\n")
```
